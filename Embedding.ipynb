```python
import torch
import torchvision
from torch.utils.data import DataLoader,Dataset
import torch.nn as nn
```

```python
transformation_operation=torchvision.transforms.Compose([torchvision.transforms.ToTensor()])
```

```python
train_dataset=torchvision.datasets.MNIST(root="./data",train=True,download=True,transform=transformation_operation)
validation_dataset=torchvision.datasets.MNIST(root="./data",download=True,train=False,transform=transformation_operation)
```

```python
train_loader=DataLoader(train_dataset,batch_size=32,shuffle=True)
validation_loader=DataLoader(validation_dataset,batch_size=32,shuffle=True)
```

```python
num_classes=10
num_channels=1
img_size=28
patch_size=7
num_patches=(img_size//patch_size)**2
emb_dim=64
attention_heads=2
transformer_blocks=4
```

```python
datapoint,label=next(iter(train_loader))
patch_embed=nn.Conv2d(num_channels,emb_dim,kernel_size=patch_size,stride=patch_size) 
print(datapoint.shape)
res=patch_embed(datapoint)
print(res.shape)
print(res.flatten(2).shape)
print(res.flatten(2).transpose(1,2).shape)
```

```python
#patch_embedding
class PatchEmbedding(nn.Module):
    def __init__(self):
        super().__init__()
        self.patch_embed=nn.Conv2d(num_channels,emb_dim,kernel_size=patch_size,stride=patch_size,padding=True) 

    def forward(self,X):
        X=self.patch_embed(X)
        X=X.flatten(2)
        X=X.transpose(1,2)
        return X
```
